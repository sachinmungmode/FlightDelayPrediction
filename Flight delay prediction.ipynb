```python
#numeric features
numeric_features = [feature for feature in flights.columns if flights[feature].dtypes!='O']
print(numeric_features)
```

```python
flights_num = flights[numeric_features]
flights_num.head()
```

```python
missing_columns = flights_num.columns

def random_imputation(df, feature):

    number_missing = df[feature].isnull().sum()
    observed_values = df.loc[df[feature].notnull(), feature]
    df.loc[df[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    
    return df
```

```python
for feature in missing_columns:
    flights_num[feature + '_imp'] = flights_num[feature]
    flights_num = random_imputation(flights_num, feature)
```

```python
from sklearn import linear_model
from sklearn.linear_model import Ridge
# linera model-ridge
deter_data = pd.DataFrame(columns = [name for name in missing_columns])

for feature in missing_columns:
        
    deter_data[feature] = flights_num[feature + "_imp"]
    parameters = list(set(flights_num.columns) - set(missing_columns) - {feature + '_imp'})
    
    #Create a Linear Regression ridge model to estimate the missing data
    model = linear_model.Ridge()
    model.fit(X = flights_num[parameters], y = flights_num[feature + '_imp'])
    
    #preserve the index of the missing data from the original dataframe
    deter_data.loc[flights_num[feature].isnull(), feature] = model.predict(flights_num[parameters])[flights_num[feature].isnull()]
```
